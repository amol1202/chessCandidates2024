### Repeating the same procedure for all the players

```python
# Constructing the URL for all the players (set the FIDE-rating >= 2700)

webID = {
         'Fabiano':'2020009',
         'Hikaru':'2016192',
         'Alireza':'12573981',
         'Nepo':'4168119',
         'Vidit':'5029465',
         'Pragg':'25059530',
         'Gukesh':'46616543',
        }

playerSeries = pd.Series(webID)
```

```python
# Heavy brute force! Looping over the hand picked URL from the web, and plotted the ratings

plt.figure(figsize=(12,4),dpi=200)
plt.ylim(2700,2810)

for i in range(len(playerSeries)):
    
    playerURL = 'https://ratings.fide.com/profile/' + playerSeries[i] + '/calculations'
    
    df_list = pd.read_html(playerURL)
    
    # Required dataframe must be holding maximum records.
    sizeInside = []
    for j in range(len(df_list)):
        sizeInside.append(len(df_list[j]))
    
    df = df_list[np.array(sizeInside).argmax()]
    
    df = df[['Period', 'RTNG']]
    df['Period'] = pd.to_datetime(df['Period'],format='%Y-%b')
    df.columns = ['Period','Rating']

    # Extracting 2023 onwards
    data_2023 = df[df['Period'] >= pd.to_datetime('2023-01-01 00:00:00')]
    
    plt.plot(data_2023['Period'], data_2023['Rating'], label = playerSeries.index[i], 
             lw = 2, marker='o', markerfacecolor = 'black', ms = 3)
    
plt.ylabel('Recent Rating Variation')
plt.xlabel('Time')
plt.legend(bbox_to_anchor = (1.04,1))
plt.set_cmap('tab10')
```

### Gathering the previous year results from chess.com <span style="color:green">&#9823;</span>

#### 2016, 2018, 2020, 2022 Results

```python
df_list = []

for year in range(2022,2014,-2):
    results= pd.read_html(str(year)+'results.html')
    results = results[0]
    results = results.drop('Player.1',axis=1)
    results.rename(columns={'#': 'Standings'}, inplace=True)
    results = results.set_index('Standings')
    rounds = results.drop(['Player','Rating','Pts.'],axis=1).columns

    results = results.fillna('-')
    
    df_list.append(results)
```

#### 2014 Results

```python
result = pd.read_html('2014results.html')
```

```python
result[0]
```

```python
result = result[0]
```

```python
# HTML tables from wiki-pages have MultiIndex columns, fetching the relevant ones.

result = result[[
        ('Player', 'Player', 'Unnamed: 1_level_2') , 
        ('Rating March 2014[8]', 'Rating March 2014[8]', 'Unnamed: 2_level_2') ,
        ('Points', 'Points', 'Unnamed: 19_level_2')
                ]]
```

```python
result.columns = ['Player', 'Rating', 'Pts.']
```

```python
# Cleaning

def convertToFloat(stringScore):
    float_score = float(stringScore[0])
    if stringScore[-1] == '½':
        float_score += 0.5
    return float_score

result['Pts.'] = result['Pts.'].apply(lambda remChar: remChar.replace('Â', '')).apply(convertToFloat)
```

```python
result
```

```python
for i in range(len(df_list)):
    mydf = df_list[i]
    mydf[mydf.columns[-8 : ]] = mydf[mydf.columns[-8 : ]].apply(lambda col: col.str.replace('Â',''))
```

```python
# 2014 Result added to the list

df_list.append(result)
```

```python
# Getting ratings vs points all years.

cols_needed = ['Rating','Pts.']

scoresDF = df_list[0][cols_needed]

for i in [1,2,3,4]:
    scoresDF = pd.concat([scoresDF, df_list[i][cols_needed]],axis=0)
```

```python
scoresDF = scoresDF.sort_values('Rating')
```

```python
# Resetting index from 0 - 1

scoresDF.index = range(len(scoresDF))
```

```python
# In case of same ratings, I take the mean pts. scored

scoresDF = scoresDF.groupby('Rating').mean()
scoresDF = scoresDF.reset_index()
```

```python
# Visualizing ratings vs the pts. scored

plt.figure(figsize=(14,4),dpi=200)
sns.barplot(x = scoresDF['Rating'], y = scoresDF['Pts.'])

plt.title('Previous Yr. Player Rating v/s Pts.scored')
```
